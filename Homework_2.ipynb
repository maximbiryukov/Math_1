```python
# Решите систему уравнений: y = x2 – 1 exp(x) + x∙(1 – y) = 1

from scipy.optimize import fsolve

x = np.linspace(-2, 3, 200)
y_1 = x**2 - 1
y_2 = (np.exp(x) + x - 1)/x
plt.plot(x, y_1)
plt.plot(x, y_2)

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x * y - 1)

x1, y1 = fsolve(equations, (-2,1))
print(x1, y1)
x2, y2 = fsolve(equations, (2,3))
print(x2, y2)
```

Повторите расчеты биномиальных коэффициентов и вероятностей k успехов в последовательности из n независимых испытаний, взяв другие значения n и k.

```python
n = [i for i in range(4,10)]
k = [i for i in range(2,5)]

for num in n:
    for num_2 in k:
        C = math.factorial(num)/(math.factorial(num_2) * math.factorial(num-num_2))
        P = C / 2**num
        print('Для {} опытов и {} попаданий вероятность равна {}'.format(num, num_2, P))
```

5. Дополните код расчетом коэффициента корреляции x и y по формуле



```python
n = 100
r = 0.8
x = np.random.rand(n)
y = r * x + (1 - r) * np.random.rand(n)
plt.scatter(x, y)
plt.grid(True)

a = (np.sum(x) * np.sum(y) - n * np.sum(x * y))/(np.sum(x) * np.sum(x) - n * np.sum(x * x))
b = (np.sum(y) - a * np.sum(x))/n

A = np.vstack([x, np.ones(len(x))]).T
a1, b1 = np.linalg.lstsq(A, y)[0]
print(a, b)
print(a1, b1)

plt.plot([0, 1], [b, a + b])
```

```python
# Дополните код расчетом коэффициента корреляции x и y по формуле

def corr_coef(x,y):
    x_m = np.mean(x)
    y_m = np.mean(y)
    upper = 0
    sum_xxmsq = 0
    sum_yymsq = 0
    
    for i in range(len(x)):
        upper += (x[i] - x_m) * (y[i] - y_m)
        sum_xxmsq += (x[i] - x_m) ** 2
        sum_yymsq += (y[i] - y_m) ** 2
        
    r = upper / np.sqrt(sum_xxmsq * sum_yymsq)
    return r

corr_coef(x, y)
```
